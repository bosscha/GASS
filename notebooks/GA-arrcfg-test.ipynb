```julia
## Create a uniform random array configuration
##

        
nant= 100
dmin= 15.0
filecfg= "random.cfg"
XX = 0
YY= 0

headerCasa= "# observatory=ALMA\n# coordsys=LOC (local tangent plane)\n# x y z diam pad#\n"
arrStr= string(headerCasa)

rng = Random.MersenneTwister(1234)
arrant= [[0.0 0.0]]
ZZ=  Random.rand!(rng, zeros(nant)) .* 2.
dist= Euclidean()

for i in 1:nant
    found=true
    while found
        XX=  rand(rng) * 500.
        YY=  rand(rng) * 500. 
        found= false
        for pt in arrant
            d= evaluate(dist, pt, [XX YY])
            if d < dmin
                found= true
            end
        end
    end
    push!(arrant, [XX YY])
end

for i in 1:nant
    arrStr *= @sprintf("%f %f %f %2.1f X%d\n", arrant[i][1], arrant[i][2], ZZ[i], 12.0, i)
end

println(arrStr)

f= open(filecfg, "w")
print(f,arrStr)
close(f)
println("## $filecfg written...")
```
